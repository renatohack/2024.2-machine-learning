### Exercício 1: Compreensão de Features no Dataset Palmer Penguins

No contexto do conjunto de dados Palmer Penguins, as `features` representam as medições físicas e características dos pinguins, como comprimento do bico, profundidade do bico e massa corporal, que são fornecidas como entrada para um modelo de machine learning. 
A relevância, qualidade, dimensionalidade, correlação entre features e normalização das features desempenham papéis críticos no desempenho do modelo. A seleção cuidadosa e a preparação delas são essenciais para garantir que o modelo possa aprender com precisão a partir dos dados e generalizar bem para novos exemplos.

### Exercício 2: Escalares, Vetores e Espaços em Machine Learning

No conjunto de dados Palmer Penguins, exemplos de escalares incluem características `individuais` como o comprimento do bico, representando valores únicos. 
Por outro lado, vetores são representados pelas `medidas físicas de um pinguim`, como comprimento do bico, profundidade do bico e comprimento da nadadeira, formando `conjuntos` de valores associados a cada pinguim. Em Machine Learning, espaços referem-se a espaços multidimensionais onde os dados são representados, com cada 'feature' correspondendo a uma dimensão. Por exemplo, o comprimento e a largura do bico de um pinguim poderiam formar um espaço bidimensional. A compreensão desses espaços é crucial para analisar a estrutura dos dados e para treinar modelos de machine learning de forma eficaz.

### Inicialização e Visualização

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import FunctionTransformer
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('penguins.csv')
df = df.drop(['year', 'sex'], axis=1)

imputer = SimpleImputer(strategy='mean')
df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']] = imputer.fit_transform(df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']])

df.head()
```

```python
variaveis_continuas = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']

num_linhas = 2
num_colunas = 2

fig, axs = plt.subplots(num_linhas, num_colunas, figsize=(12, 8))
axs = axs.flatten()

for i, variavel in enumerate(variaveis_continuas):
    sns.histplot(data=df, x=variavel, kde=True, ax=axs[i])  # Cria um histograma com densidade
    axs[i].set_title(f'Distribuição de {variavel}')  # Adiciona título ao subplot
    axs[i].set_xlabel(variavel)  # Adiciona rótulo ao eixo x
    axs[i].set_ylabel('Frequência')  # Adiciona rótulo ao eixo y

plt.tight_layout()
plt.show()
```

### Exercício 3: Discretização com Bins Fixos no Dataset Palmer Penguins

**Utilize a técnica de quantização com bins fixos para discretizar uma variável contínua do dataset Palmer Penguins. Explique a razão pela qual você escolheu essa variável e como a discretização pode afetar a análise.**

A razão pela qual escolhi a variável de massa corporal é porque as outras features possuiam um intervalo muito limitado, ao passo que os valores iniciais do intervalo da feature de massa corporal são praticamente metade dos valores finais, há uma variância muito grande.

```python
labels = ['light', 'normal', 'heavy']
df['body_mass_fixed_bins'] = pd.cut(df['body_mass_g'], bins=3, labels=labels)

print(df[['body_mass_g', 'body_mass_fixed_bins']].head())
```

### Exercício 4:Discretização com Bins Variáveis no Dataset Palmer Penguins

**Aplique a técnica de quantização com bins variáveis em uma variável contínua do dataset Palmer Penguins. Compare os resultados com a discretização feita no exercício anterior.**

Como podemos notar nos resultados abaixo, o número de indivíduos em cada grupo muda bruscamente dependendo da técnica utilizada. Praticamente 50% dos pinguins seriam considerados "leves", enquanto aproximadamente 20% seria pesado, na estratégia de bin fixo.

```python
quantis = df['body_mass_g'].quantile([0, 0.33, 0.66, 1]).tolist()
labels = ['light', 'normal', 'heavy']
df['body_mass_var_bins'] = pd.cut(df['body_mass_g'], bins=quantis, labels=labels, include_lowest=True)

print(df[['body_mass_g', 'body_mass_var_bins']].head())
```

```python
num_linhas = 1
num_colunas = 2

fig, axs = plt.subplots(num_linhas, num_colunas, figsize=(12, 8))
axs = axs.flatten()

sns.histplot(data=df, x='body_mass_fixed_bins', ax=axs[0])  # Cria um histograma com densidade
axs[1].set_title(f'Distribuição de body_mass_fixed_bins')  # Adiciona título ao subplot
axs[1].set_xlabel('body_mass_fixed_bins')  # Adiciona rótulo ao eixo x
axs[1].set_ylabel('Frequência')  # Adiciona rótulo ao eixo y

sns.histplot(data=df, x='body_mass_var_bins', ax=axs[1])  # Cria um histograma com densidade
axs[1].set_title(f'Distribuição de body_mass_var_bins')  # Adiciona título ao subplot
axs[1].set_xlabel('body_mass_var_bins')  # Adiciona rótulo ao eixo x
axs[1].set_ylabel('Frequência')  # Adiciona rótulo ao eixo y

plt.tight_layout()
plt.show()
```

### Exercício 5: Uso de FunctionTransformer no Dataset Palmer Penguins

**Aplique a FunctionTransformer do Scikit-Learn em uma variável do dataset Palmer Penguins. Descreva o processo e explique como essa transformação pode ser benéfica.**

A classe FunctionTransformer do Scikit-Learn permite aplicar uma função arbitrária a cada elemento de uma matriz ou DataFrame. Isso é útil quando você deseja aplicar uma transformação personalizada aos seus dados.

A aplicação da classe FunctionTransformer pode ser altamente benéfica por ser possível realizar pré-processamentos personalizados através dessa classe. Além disso, na engenharia de recursos, a FunctionTransformer possibilita a criação de novas características a partir de características existentes, empregando funções matemáticas ou lógicas mais complexas. Outra vantagem é a capacidade de implementar e aplicar normalizações especializadas que não estão disponíveis nos transformadores padrão do Scikit-Learn, oferecendo uma maior flexibilidade e adaptabilidade aos dados e às necessidades específicas do problema em questão.

```python
def bill_pow2(x):
    return x * x 

transformer = FunctionTransformer(func=bill_pow2)
df['bill_length_mm_pow2'] = transformer.transform(df['bill_length_mm'])

print(df.head())
```

### Exercício 6: Aplicação de PowerTransformer no Dataset Palmer Penguins

**Utilize a PowerTransformer do Scikit-Learn para transformar uma variável do dataset Palmer Penguins. Discuta os benefícios dessa transformação.**

A classe PowerTransformer do Scikit-Learn é usada para aplicar transformações de potência aos dados, sendo útil para tornar a distribuição dos dados mais gaussiana. Isso pode ser particularmente benéfico em casos onde os dados não seguem uma distribuição normal, o que é uma premissa comum em muitos modelos de machine learning.

```python
transformador = PowerTransformer()

df['body_mass_g_transform'] = transformador.fit_transform(df[['body_mass_g']])

variaveis = ['body_mass_g', 'body_mass_g_transform']

num_linhas = 1
num_colunas = 2

fig, axs = plt.subplots(num_linhas, num_colunas, figsize=(12, 8))
axs = axs.flatten()

for i, variavel in enumerate(variaveis):
    sns.histplot(data=df, x=variavel, kde=True, ax=axs[i])  # Cria um histograma com densidade
    axs[i].set_title(f'Distribuição de {variavel}')  # Adiciona título ao subplot
    axs[i].set_xlabel(variavel)  # Adiciona rótulo ao eixo x
    axs[i].set_ylabel('Frequência')  # Adiciona rótulo ao eixo y

plt.tight_layout()
plt.show()
```

### Exercício 7:Normalização Min-Max no Dataset Palmer Penguins

**Aplique a normalização Min-Max do Scikit-Learn em uma ou mais variáveis do dataset Palmer Penguins. Explique como essa normalização impacta o modelo de Machine Learning.**

A normalização Min-Max reduz a sensibilidade a unidades, pois ajusta os valores das variáveis para um intervalo específico, tornando o modelo menos suscetível a variações nas unidades das variáveis e melhorando sua estabilidade. 
A normalização Min-Max também preserva as relações de ordem entre os dados, garantindo que a ordem dos valores antes e depois da normalização permaneça a mesma, o que é fundamental para o aprendizado correto do modelo.

```python
scaler = MinMaxScaler()
df[['bill_depth_mm_min_max']] = scaler.fit_transform(df[['bill_depth_mm']])

print(df.head())
```

### Exercício 8: Normalização com Standard Scaler no Dataset Palmer Penguins

**Utilize o StandardScaler do Scikit-Learn para normalizar variáveis do dataset Palmer Penguins. Discuta a diferença entre esta técnica e a normalização Min-Max.**

O `StandardScaler` do Scikit-Learn padroniza os dados removendo a média e escalonando para uma unidade de desvio padrão, resultando em dados com média zero e desvio padrão igual a um. Essa técnica é sensível à distribuição dos dados e é mais adequada para dados com distribuição gaussiana. Por outro lado, o `MinMaxScaler` dimensiona os dados para um intervalo específico, geralmente entre 0 e 1, preservando as relações de ordem entre os dados e sendo menos sensível à distribuição dos dados. Enquanto o `StandardScaler` é útil quando a escala das variáveis é conhecida e importante, o `MinMaxScaler` é mais apropriado quando a distribuição dos dados é desconhecida ou quando é crucial preservar as relações de ordem entre os dados.

```python
scaler = StandardScaler()
df[['bill_length_mm_stdscaler', 'bill_depth_mm_stdscaler', 'flipper_length_mm_stdscaler', 'body_mass_g_stdscaler']] = scaler.fit_transform(df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']])

print(df.head())
```
