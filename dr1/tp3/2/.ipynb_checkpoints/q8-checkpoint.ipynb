```python
import pandas as pd
```

```python
df = pd.read_csv('adult.csv')
df = df[['age', 'workclass']]
df.head()
```

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()

labels = lb.fit_transform(df['workclass'])
temp_df = pd.DataFrame(labels)

df_encoded = df.join(temp_df)
df_encoded.head()
```
