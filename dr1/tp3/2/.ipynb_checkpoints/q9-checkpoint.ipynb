```python
import pandas as pd
```

```python
df = pd.read_csv('adult.csv')
df = df[['age', 'workclass']]
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

oh = OneHotEncoder(sparse_output=False).set_output(transform='pandas')
oh_encoded = oh.fit_transform(df[['workclass']])

df = pd.concat([df, oh_encoded], axis =1).drop(columns=['workclass'])
df.head()
```
