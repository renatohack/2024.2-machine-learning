```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
import plotly.graph_objs as go
from sklearn.cluster import KMeans
import warnings
import os
from sklearn.preprocessing import StandardScaler
from yellowbrick.cluster import KElbowVisualizer
from sklearn.decomposition import PCA
from sklearn import metrics
warnings.filterwarnings("ignore")
py.offline.init_notebook_mode(connected = True)
```

### Separe as colunas que considera relevantes para o agrupamento de dados.

```python
ds = pd.read_csv('paises.csv')
ds = ds.drop(['country'], axis=1)
ds.head()
```

```python
plt.style.use('fivethirtyeight')
```

```python
numeric = ['child_mort', 'exports', 'health', 'imports', 'income', 'inflation', 'life_expec', 'total_fer', 'gdpp']

plt.figure(1 , figsize = (15 , 15))
n = 0
for x in numeric:
    n += 1
    plt.subplot(3 , 3 , n)
    plt.subplots_adjust(hspace =0.5 , wspace = 0.5)
    sns.distplot(ds[x] , bins = 20)
    plt.title('Distplot of {}'.format(x))
plt.show()
```

```python
plt.figure(1 , figsize = (30, 30))
n = 0
for x in numeric:
    for y in numeric:
        n += 1
        plt.subplot(9 , 9 , n)
        plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
        sns.regplot(x = x , y = y , data = ds)
        plt.ylabel(y.split()[0]+' '+y.split()[1] if len(y.split()) > 1 else y )
plt.show()
```

### Trate e normalize os dados. (Std Scaler e PCA)

```python
scaler = StandardScaler()
ds[numeric] = scaler.fit_transform(ds[numeric])
pca = PCA(n_components=2)
ds_pca = pd.DataFrame(pca.fit_transform(ds), columns=['0', '1'])

ds_pca.head()
```

### Utilize as Inrcias para tentar visualizar o cotovelo. Utilize o yellowbrick KElbowVisualizer para procurar o cotovelo. Qual valor de K voc utilizar?

```python
X1 = ds_pca[['0', '1']].iloc[:,:].values
inercias = []
valores_k = []
for n in range(1,15):
  kmeans = KMeans(n_clusters=n, random_state=0).fit(X1)
  valores_k.append(n)
  inercias.append(kmeans.inertia_)
```

```python
plt.figure(1, figsize=(15,6))
plt.plot(valores_k, inercias,'o')
plt.plot(valores_k, inercias,'-', alpha=0.5)
plt.xlabel('N de Clusters'), plt.ylabel("Inercia")
plt.show()
```

```python
model = KMeans(n_clusters=15, random_state=0)
visualizer = KElbowVisualizer(model,k=15)
visualizer.fit(X1)
visualizer.show()
```

### Plote os dados em um scatterplot.

```python
kmeans = KMeans(n_clusters=4, random_state=0).fit(X1)
centroids = kmeans.cluster_centers_
```

```python
h = 0.02
x_min, x_max = X1[:, 0].min() - 1, X1[:, 0].max() + 1
y_min, y_max = X1[:, 1].min() - 1, X1[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
plt.figure(1 , figsize = (15 , 7) )
plt.clf()
Z = Z.reshape(xx.shape)
plt.imshow(Z , interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap = plt.cm.Pastel2, aspect = 'auto', origin='lower')

plt.scatter( x = '0', y = '1' , data = ds_pca, c = kmeans.labels_ , s = 200 )
plt.scatter(x = centroids[: , 0] , y =  centroids[: , 1] , s = 300 , c = 'red' , alpha = 0.5)
plt.ylabel('F2') , plt.xlabel('F1')
plt.show()
```

### Utilize o Silhouette Score para avaliar o modelo. O que o Silhouette Score avalia como um bom modelo? Voc poder utilizar o ndice de Rand para avaliar o modelo?
O Silhouette Score  uma mtrica usada para avaliar a qualidade de clusters em algoritmos de agrupamento, como K-Means. 

Ele mede o quo bem cada ponto de dado est agrupado com outros pontos que so semelhantes a ele e o quo distinto ele  de outros clusters.

O ndice de Rand no pode ser utilizado diretamente pois requer uma referncia para comparar os agrupamentos gerados pelo modelo, ento, sem essa classificao verdadeira, no h como calcular a mtrica de similaridade entre dois conjuntos de agrupamentos.

```python
score = metrics.silhouette_score(X1, kmeans.labels_)
score
```

### Voc faria algo para melhorar a performance do modelo?
### Em quais casos no  aconselhvel utilizar K-means?
Para melhorar a performance do K-Means, pode-se otimizar o nmero de clusters usando mtodos como o cotovelo ou o Silhouette Score, padronizar os dados para garantir que todas as variveis contribuam igualmente, e utilizar a inicializao K-Means++ para evitar convergncia em mnimos locais. Outras estratgias incluem a remoo de outliers, ajuste de hiperparmetros, e a reduo de dimensionalidade com tcnicas como PCA. Essas prticas ajudam a obter clusters mais precisos e eficientes.

O K-Means no  aconselhvel em situaes onde os clusters no so esfricos, tm tamanhos muito diferentes, ou se sobrepem, alm de ser sensvel a outliers e pouco eficiente em dados de alta dimensionalidade ou com variveis em escalas diferentes. Nesses casos,  mais apropriado considerar outros algoritmos de clustering, como DBSCAN ou Gaussian Mixture Models, que lidam melhor com essas limitaes.
