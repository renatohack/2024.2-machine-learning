### Init

```python
from sklearn.datasets import load_diabetes
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeRegressor
```

```python
diabetes = load_diabetes()
X = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
y = pd.Series(diabetes.target)

X.head(), y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Training

```python
dt = DecisionTreeRegressor(random_state=42)

param_grid = {
    'max_depth': range(1, 10, 1),
    'min_samples_leaf': range(1, 10, 1),
    'max_leaf_nodes': [None, 10, 20, 30],
}

grid_dt = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_dt.fit(X_train, y_train)
```

```python
best_params = grid_dt.best_params_
best_params
```

### Scoring

```python
y_pred = grid_dt.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

mse, r2
```

### Visualize

```python
model = grid_dt.best_estimator_

plt.figure(figsize=(20,10))
plot_tree(model, feature_names=diabetes.feature_names, filled=True)
plt.title("Árvore de Decisão - Diabetes Dataset")
plt.show()
```

```python
importances = model.feature_importances_

plt.figure(figsize=(10,5))
plt.barh(diabetes.feature_names, importances)
plt.xlabel('Importância')
plt.ylabel('Features')
plt.title('Importância das Features - Decision Tree Regressor')
plt.show()
```

```python
X_reduced = X[['bmi', 'bp']]

X_train_reduced, X_test_reduced, y_train_reduced, y_test_reduced = train_test_split(X_reduced, y, test_size=0.3, random_state=42)

model_reduced = DecisionTreeRegressor(random_state=42)
model_reduced.fit(X_train_reduced, y_train_reduced)

x_min, x_max = X_reduced.iloc[:, 0].min() - 0.05, X_reduced.iloc[:, 0].max() + 0.05
y_min, y_max = X_reduced.iloc[:, 1].min() - 0.05, X_reduced.iloc[:, 1].max() + 0.05
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

Z = model_reduced.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X_reduced.iloc[:, 0], X_reduced.iloc[:, 1], c=y, edgecolor='k', s=20, cmap='viridis')
plt.colorbar()
plt.xlabel('BMI')
plt.ylabel('BP')
plt.title('Previsões - Decision Tree Regressor')
plt.show()
```

### Testing different Params

```python
best_params = grid_dt.best_params_
best_params
```

#### max_depth

```python
max_depth = 4
max_leaf_nodes = best_params['max_leaf_nodes']
min_samples_leaf = int(best_params['min_samples_leaf'])

dt = DecisionTreeRegressor(random_state=42,
                            max_depth=max_depth, 
                            max_leaf_nodes=max_leaf_nodes, 
                            min_samples_leaf=min_samples_leaf)

dt.fit(X_train, y_train)

y_pred = dt.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('MSE:', mse, 'R2:', r2)

model = dt

plt.figure(figsize=(20,10))
plot_tree(model, feature_names=diabetes.feature_names, filled=True)
plt.title("Árvore de Decisão - Diabetes Dataset")
plt.show()

importances = model.feature_importances_

plt.figure(figsize=(10,5))
plt.barh(diabetes.feature_names, importances)
plt.xlabel('Importância')
plt.ylabel('Features')
plt.title('Importância das Features - Decision Tree Regressor')
plt.show()
```
