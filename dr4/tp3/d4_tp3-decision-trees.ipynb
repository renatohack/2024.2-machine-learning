# Init

```python
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, cross_val_score
import numpy as np
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
data = load_breast_cancer()
X, y = data.data, data.target
```

# Processing - Árvores Individuais

## Decision Tree - Max Depth

```python
param_grid_depth = {'max_depth': np.arange(1, 11)}
model_depth = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid_depth, cv=5)
model_depth.fit(X, y)
```

## Decision Tree - Max Leaf Nodes

```python
param_grid_leaf_nodes = {'max_leaf_nodes': np.arange(2, 21)}
model_leaf_nodes = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid_leaf_nodes, cv=5)
model_leaf_nodes.fit(X, y)
```

## Decision Tree - Min Samples Leaf

```python
param_grid_samples_leaf = {'min_samples_leaf': np.arange(1, 11)}
model_samples_leaf = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid_samples_leaf, cv=5)
model_samples_leaf.fit(X, y)
```

# Resultados Árvores individuais

## Melhores params

```python
# Resultados dos melhores parâmetros e acurácia de validação cruzada para cada modelo
best_depth = model_depth.best_params_['max_depth']
best_leaf_nodes = model_leaf_nodes.best_params_['max_leaf_nodes']
best_samples_leaf = model_samples_leaf.best_params_['min_samples_leaf']

# Avaliação de desempenho
scores_depth = cross_val_score(model_depth.best_estimator_, X, y, cv=5)
scores_leaf_nodes = cross_val_score(model_leaf_nodes.best_estimator_, X, y, cv=5)
scores_samples_leaf = cross_val_score(model_samples_leaf.best_estimator_, X, y, cv=5)

{
    "best_depth": best_depth,
    "scores_depth": scores_depth.mean(),
    "best_leaf_nodes": best_leaf_nodes,
    "scores_leaf_nodes": scores_leaf_nodes.mean(),
    "best_samples_leaf": best_samples_leaf,
    "scores_samples_leaf": scores_samples_leaf.mean(),
}
```

## Features Importance

```python
# Importância das features
importances_depth = model_depth.best_estimator_.feature_importances_
importances_leaf_nodes = model_leaf_nodes.best_estimator_.feature_importances_
importances_samples_leaf = model_samples_leaf.best_estimator_.feature_importances_

# Plotando a importância das features
features = data.feature_names
fig, axs = plt.subplots(3, 1, figsize=(12, 15))
fig.tight_layout(pad=6.0)

# Modelo 1: max_depth
sns.barplot(x=importances_depth, y=features, ax=axs[0])
axs[0].set_title(f'Feature Importance - max_depth (Best: {model_depth.best_params_["max_depth"]})')

# Modelo 2: max_leaf_nodes
sns.barplot(x=importances_leaf_nodes, y=features, ax=axs[1])
axs[1].set_title(f'Feature Importance - max_leaf_nodes (Best: {model_leaf_nodes.best_params_["max_leaf_nodes"]})')

# Modelo 3: min_samples_leaf
sns.barplot(x=importances_samples_leaf, y=features, ax=axs[2])
axs[2].set_title(f'Feature Importance - min_samples_leaf (Best: {model_samples_leaf.best_params_["min_samples_leaf"]})')

plt.show()
```

## Métricas

```python
y_pred_depth = model_depth.best_estimator_.predict(X)
y_pred_leaf_nodes = model_leaf_nodes.best_estimator_.predict(X)
y_pred_samples_leaf = model_samples_leaf.best_estimator_.predict(X)

print("Classification Report - max_depth")
print(classification_report(y, y_pred_depth, target_names=data.target_names))
print('---')

print("Classification Report - max_leaf_nodes")
print(classification_report(y, y_pred_leaf_nodes, target_names=data.target_names))
print('---')

print("Classification Report - min_samples_leaf")
print(classification_report(y, y_pred_samples_leaf, target_names=data.target_names))
```

## Matriz de Confusão

```python
fig, axs = plt.subplots(3, 1, figsize=(10, 15))
fig.tight_layout(pad=6.0)

# Modelo 1: max_depth
ConfusionMatrixDisplay.from_estimator(model_depth.best_estimator_, X, y, ax=axs[0], display_labels=data.target_names, cmap='Blues')
axs[0].set_title('Confusion Matrix - max_depth')

# Modelo 2: max_leaf_nodes
ConfusionMatrixDisplay.from_estimator(model_leaf_nodes.best_estimator_, X, y, ax=axs[1], display_labels=data.target_names, cmap='Blues')
axs[1].set_title('Confusion Matrix - max_leaf_nodes')

# Modelo 3: min_samples_leaf
ConfusionMatrixDisplay.from_estimator(model_samples_leaf.best_estimator_, X, y, ax=axs[2], display_labels=data.target_names, cmap='Blues')
axs[2].set_title('Confusion Matrix - min_samples_leaf')

plt.show()
```

# Processing - Combinando Modelos

```python
# Melhores parâmetros ajustados

best_model = DecisionTreeClassifier(
    max_depth=model_depth.best_params_['max_depth'],
    max_leaf_nodes=model_leaf_nodes.best_params_['max_leaf_nodes'],
    min_samples_leaf=model_samples_leaf.best_params_['min_samples_leaf'],
    random_state=42
)
```

## Bagging

```python
bagging_model = BaggingClassifier(estimator=best_model, n_estimators=10, random_state=42)
bagging_model.fit(X, y)
```

## Random Forest

```python
random_forest_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=model_depth.best_params_['max_depth'],
    max_leaf_nodes=model_leaf_nodes.best_params_['max_leaf_nodes'],
    min_samples_leaf=model_samples_leaf.best_params_['min_samples_leaf'],
    random_state=42
)
random_forest_model.fit(X, y)
```

## Boosting

```python
boosting_model = AdaBoostClassifier(estimator=best_model, n_estimators=50, algorithm='SAMME', random_state=42)
boosting_model.fit(X, y)
```

## Voting Classifier

```python
voting_model = VotingClassifier(estimators=[
    ('depth', model_depth.best_estimator_),
    ('leaf_nodes', model_leaf_nodes.best_estimator_),
    ('samples_leaf', model_samples_leaf.best_estimator_),
    ('bagging', bagging_model),
    ('random_forest', random_forest_model),
    ('boosting', boosting_model)
], voting='soft')

voting_model.fit(X, y)
```

# Resultados Ensembles

## Melhores params

```python
scores_bagging = cross_val_score(bagging_model, X, y, cv=5)
scores_random_forest = cross_val_score(random_forest_model, X, y, cv=5)
scores_boosting = cross_val_score(boosting_model, X, y, cv=5)
scores_voting = cross_val_score(voting_model, X, y, cv=5)

{
    "scores_bagging": scores_bagging.mean(),
    "scores_random_forest": scores_random_forest.mean(),
    "scores_boosting": scores_boosting.mean(),
    "scores_voting": scores_voting.mean(),
}
```

## Features Importance

## Métricas

```python
y_pred_bagging = bagging_model.predict(X)
y_pred_random_forest = random_forest_model.predict(X)
y_pred_boosting = boosting_model.predict(X)
y_pred_voting = voting_model.predict(X)

print("Classification Report - Bagging")
print(classification_report(y, y_pred_bagging, target_names=data.target_names))
print()
print()

print("Classification Report - Random Forest")
print(classification_report(y, y_pred_random_forest, target_names=data.target_names))
print()
print()

print("Classification Report - Boosting")
print(classification_report(y, y_pred_boosting, target_names=data.target_names))
print()
print()

print("Classification Report - Voting Classifier")
print(classification_report(y, y_pred_voting, target_names=data.target_names))
```

## Matriz de Confusão

```python
fig, axs = plt.subplots(4, 1, figsize=(10, 15))
fig.tight_layout(pad=6.0)

ConfusionMatrixDisplay.from_estimator(bagging_model, X, y, ax=axs[0], display_labels=data.target_names, cmap='Blues')
axs[0].set_title('Confusion Matrix - Bagging')

ConfusionMatrixDisplay.from_estimator(random_forest_model, X, y, ax=axs[1], display_labels=data.target_names, cmap='Blues')
axs[1].set_title('Confusion Matrix - Random Forest')

ConfusionMatrixDisplay.from_estimator(boosting_model, X, y, ax=axs[2], display_labels=data.target_names, cmap='Blues')
axs[2].set_title('Confusion Matrix - Boosting')

ConfusionMatrixDisplay.from_estimator(voting_model, X, y, ax=axs[3], display_labels=data.target_names, cmap='Blues')
axs[3].set_title('Confusion Matrix - Voting Classifier')

plt.show()
```
